```python
import pandas as pd
import seaborn as sns
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
base_path = Path("./results/")
trivial_loop_df = pd.read_csv(base_path / "trivial_loop.csv") 
non_trivial_loop_df = pd.read_csv(base_path / "non_trivial_loop.csv") 
merge_sort_df = pd.read_csv(base_path / "merge_sort.csv") 
bubble_sort_df = pd.read_csv(base_path / "bubble_sort.csv") 
quick_sort_df = pd.read_csv(base_path / "quick_sort.csv") 
```

```python
sns.regplot(data=trivial_loop_df, x="iteration", y="execution_time_s")
plt.xlabel("Number of iterations")
plt.ylabel("Execution Time (s)")
plt.title("Run Time vs Loop Size for Trivial Loop")
```

```python
sns.regplot(data=non_trivial_loop_df, x="iteration", y="execution_time_s")
plt.xlabel("Number of iterations")
plt.ylabel("Execution Time (s)")
plt.title("Run Time vs Loop Size for Non-Trivial Loop")
```

```python
combined_df = merge_sort_df.rename(mapper={"execution_time_s": "merge_sort_s"}, axis=1)
combined_df["bubble_sort_s"] = bubble_sort_df["execution_time_s"]
combined_df["quick_sort_s"] = quick_sort_df["execution_time_s"]
combined_df = combined_df.set_index("iteration")
sns.lineplot(data=combined_df)
plt.xlabel("Number of iterations")
plt.ylabel("Execution Time (s)")
plt.title("Run Time in Seconds vs Loop Size for sorting algorithms")
_ = plt.legend(labels=["Merge Sort","Bubble Sort", "Quick Sort"])
```

Looks like quick sort was the fastest because its average time performance is slightly better than merge sort's despite having a worse worst case complexity. Gotta love that referential locality
