```python
import pandas as pd
import seaborn as sns
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
base_path = Path("./results/")
trivial_loop_df = pd.read_csv(base_path / "trivial_loop.csv") 
non_trivial_loop_df = pd.read_csv(base_path / "non_trivial_loop.csv") 
```

```python
sns.scatterplot(data=trivial_loop_df, x="iteration", y="execution_time_s")
plt.xlabel("Number of iterations")
plt.ylabel("Execution Time (s)")
plt.title("Run Time vs Loop Size for Trivial Loop")
```

```python
sns.scatterplot(data=non_trivial_loop_df, x="iteration", y="execution_time_s")
plt.xlabel("Number of iterations")
plt.ylabel("Execution Time (s)")
plt.title("Run Time vs Loop Size for Non-Trivial Loop")
```
